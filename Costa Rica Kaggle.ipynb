Now we have a list of the most influencing factors on a person's target.  Lets see if they can give us a good prediction using a basic TF estimator DNN.  Lets see what these also actually mean - 

hogar_nin = number of children 0 - 19 in household

r4t1 = persons younger than 12 years of age
these two beneath can be dropped as they are captured by r4t1 better

r4m1 = female younger than 12 years old

r4h1 = male younger than 12 years old

eviv1 = if floors are bad

piscocemento = if floor is cement

epared1 = if walls are bad

rooms = number of all rooms (the more rooms, the less likely to be economically in need)

v18q = owns a tablet

paredblolad = if predominant material on the outside wall is block or brick

escolari = years of schooling

pisomoscer = if predominant material on the floor is mosaic,  ceramic,  terrazo"

etecho3 = if roof are good

epared3 = if walls are good

eviv3 = if floors are good

cielorazo = if the house has a ceiling

```python
import tensorflow as tf
```

```python
cols_to_use = []
```

```python
most_important_coerr = filter(lambda x : abs(x[1])>.2, sorted_coeff)
```

```python
for k,v in list(most_important_coerr):
    if(k!='r4m1' and k!='r4h`'):
        cols_to_use.append(k)
```

```python
cols_to_use
```

```python
#Niaveley make feature columns, can make this better later
kids_under_12 = tf.feature_column.numeric_column('r4h1')
cement_floors = tf.feature_column.numeric_column('pisocemento')
num_rooms = tf.feature_column.numeric_column('rooms')
tablet = tf.feature_column.numeric_column('v18q')
roof_bad = tf.feature_column.numeric_column('etecho3')
walls_bad = tf.feature_column.numeric_column('epared3')
schooling = tf.feature_column.numeric_column('escolari')
cieling = tf.feature_column.numeric_column('cielorazo')
num_children = tf.feature_column.numeric_column('hogar_nin')
bad_floors = tf.feature_column.numeric_column('eviv1')
walls_good = tf.feature_column.numeric_column('epared1')
block_wall = tf.feature_column.numeric_column('paredblolad')
mosaic_floor = tf.feature_column.numeric_column('pisomoscer')
good_floors = tf.feature_column.numeric_column('eviv3')
```

```python
x_data = train[cols_to_use]
```

```python
#Need to do this because DNN estimator does a classifier starting at 0 to n-1, not 1 to n
labels = train['Target']-1
```

```python
labels
```

```python
feat_cols = [kids_under_12, cement_floors, num_rooms, tablet, roof_bad, walls_bad, schooling, cieling, num_children, bad_floors, walls_good, block_wall, mosaic_floor, good_floors]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, labels, test_size=.1)
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train, y=y_train,
                                                 batch_size=8,num_epochs=1000,
                                                 shuffle=True)
```

```python
dnn_model = tf.estimator.DNNClassifier(hidden_units=[15,15,15,10,5,5],feature_columns=feat_cols,n_classes=4)
```

```python
dnn_model.train(input_fn=input_func,steps=10000)
```

```python
eval_input_func  = tf.estimator.inputs.pandas_input_fn(x=X_test,y=y_test,
                                                      batch_size=10,num_epochs=1,
                                                      shuffle=False)
```

```python
dnn_model.evaluate(eval_input_func)
```
